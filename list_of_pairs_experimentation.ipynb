```python
def get_all_possible_pairs(input_list): 
    
    print(f"item count: {len(input_list)}")
    
    all_possible_pairs = set()
    k = 1 # when k is the len of the list each item will be paired with itself 
    while k < len(input_list):
        for i in range(0, len(input_list)):
            # print("\n")
            next_item = i+k # k is the distance between the current item and the next item
            
            # print("next_item:", next_item)

            next_item_fixed = next_item % len(input_list)
            # print("next_item_fixed:", next_item_fixed)

            # print(i, next_item_fixed)

            # print(input_list[i], input_list[next_item_fixed])
            
            pair_tup = (input_list[i], input_list[next_item_fixed])
            
            # don't add reverse version of existing tuple 
            
            if pair_tup not in all_possible_pairs and pair_tup[::-1] not in all_possible_pairs:

                all_possible_pairs.add(pair_tup)

        k += 1 
        
    print(f"possible pairs: {len(all_possible_pairs)} ")
    
    print("pairs/total items", len(all_possible_pairs) / len(input_list))
    
    import re
    # make set into string then count occurances of A
    appposs_str = re.sub(r'\W+', '', str(all_possible_pairs))

    A_occurances = appposs_str.count("A")

    print(f"item occurances:", A_occurances)
    
    return(all_possible_pairs)

input_list_4 = ['A', 'B', 'C', 'D']
input_list_5 = ['A', 'B', 'C', 'D', 'E']    
input_list_6 = ['A', 'B', 'C', 'D', 'E', 'F']
input_list_7 = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
input_list_8 = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']

all_pairs_4 = get_all_possible_pairs(input_list_4)
print(all_pairs_4, "\n")

all_pairs_5 = get_all_possible_pairs(input_list_5)
print(all_pairs_5, "\n")

all_pairs_6 = get_all_possible_pairs(input_list_6)
print(all_pairs_6, "\n")

all_pairs_7 = get_all_possible_pairs(input_list_7)
print(all_pairs_7, "\n")

all_pairs_8 = get_all_possible_pairs(input_list_8)
print(all_pairs_8, "\n")
```

```python
def make_dict_items(list_of_items, list_of_pairs):
    print("\n\n")
    items_dict = {}
    
    for item in list_of_items:
        # loop over letters
        for pair in list_of_pairs:
            
            if item in pair:
                if item not in items_dict:
                    items_dict[item] = []

                items_dict[item].append(pair)
    
    return items_dict

dict4 = make_dict_items(input_list_4, all_pairs_4)
print(dict4)
dict5 = make_dict_items(input_list_5, all_pairs_5)
print(dict5)
dict6 = make_dict_items(input_list_6, all_pairs_6)
print(dict6)
dict7 = make_dict_items(input_list_7, all_pairs_7)
print(dict7)
dict8 = make_dict_items(input_list_8, all_pairs_8)
print(dict8)
```
