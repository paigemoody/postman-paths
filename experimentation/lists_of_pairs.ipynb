```python
# recursive solution: https://stackoverflow.com/questions/5360220/how-to-split-a-list-into-pairs-in-all-possible-ways

import itertools

alpha_list = ['A','B','C','D', 'E', 'F', 'G', 'H']

# from: https://stackoverflow.com/questions/5360220/how-to-split-a-list-into-pairs-in-all-possible-ways
output_list_of_lists = [] 

combos = []

def all_pairs(input_lst):
    
    # print("combos:",combos)
    
    
    # print("---")
    
    
    # base case - when there are no items remaining, or one item 
    if len(input_lst) < 2:
        
        #print("\n!!!end - hit base case")
        # print("\ninput_lst:",input_lst)
        # print("-----------\n\n\n\n\n")
        return []
        
    else: 

        # print("\ninput_lst:",input_lst)

        first_item = input_lst[0] # first item in list 
        # print(" first_item:",first_item)

        # loop over every item in list
        for i in range(1,len(input_lst)):
            
            # print("\n  FOR LOOP i:", i)

            pair = (first_item,input_lst[i])
            
            # print("\n  pair:", pair)
            
            combos.append(pair)
            
            other_items_list = input_lst[1:i] + input_lst[i+1:] 

            # print("\n  other items:", other_items_list)

            # for rest in all_pairs(input_lst[1:i]+input_lst[i+1:]):
                
            for rest in all_pairs(other_items_list):
                
                # print("rest:", rest)
                
                # print("[pair] + rest", [pair] + rest)

                # yield [pair] + rest
    
                return [pair] + rest
    
    # print("output_list_of_lists",output_list_of_lists)
    
    # print("len input list:", len(alpha_list))
    
    if len(combos) >= len(alpha_list)/2: 
        print(combos)
        output_list_of_lists.append(combos)
    
    # reset combos vairable 
    del combos[0:]
    
    return []

    
all_pairs(alpha_list)

```

```python
def get_list_of_all_pairs_lists(lst):
    
    list_of_possible_pairs_lists = []
    all_pairs_object = all_pairs(lst)
    
    for pairs_list in all_pairs_object: 
        list_of_possible_pairs_lists.append(pairs_list)
    
    return list_of_possible_pairs_lists
    
# possible_pairings_list = get_list_of_all_pairs_lists(alpha_list)

possible_pairings_list = get_list_of_all_pairs_lists(alpha_list)

print(possible_pairings_list)

for pair_list in possible_pairings_list:
    print()
    print(pair_list)
```

```python
# updated variables

alpha_list = ['A','B','C','D', 'E', 'F']

def all_pairs(lst):
    
    if len(lst) < 2:
        # if there are zero items or just one 
        # return (yield?) an empty list
        yield []
        return
    
    if len(lst) % 2 == 1:
        # Handle odd length list
        for i in range(len(lst)):
            
            # items from 0 to i 
            items_to_left_ = lst[:i]    

            # items right of i to end of list
            items_to_right_ = lst[i+1:]
            
            for result in all_pairs(items_to_left_ + items_to_right_):
                yield result
    else:
        cur_first_item = lst[0]
        
        for i in range(1,len(lst)):
            
            next_item = lst[i]
            
            pair = (cur_first_item, next_item)
            
            items_to_left_of_pair = lst[1:i]
            items_to_right_of_pair = lst[i+1:]
            
            for rest in all_pairs(items_to_left_of_pair + items_to_right_of_pair):
                yield [pair] + rest

for x in all_pairs(alpha_list):
    print(x)
    
```
